```python
# Load necessary libraries
import numpy as np

# Read the text file
text_file = open('alice_in_wonderland.txt', 'r')
text = text_file.read()
text_file.close()

# Display length of the text
print(f'Text length: {len(text)} characters')
```

```python
# Convert text to lowercase
text = text.lower()

# Example: Remove new lines and replace with space
text = text.replace('\n', ' ')
```

```python
# Create a mapping from characters to integers
chars = sorted(list(set(text)))
char_to_index = {c: i for i, c in enumerate(chars)}
index_to_char = {i: c for i, c in enumerate(chars)}

# Display the number of unique characters
print(f'Unique characters: {len(chars)}')
```

```python
# Define sequence length
seq_length = 100
step = 1

# Prepare input and output sequences
sequences = []
next_chars = []

for i in range(0, len(text) - seq_length, step):
    sequences.append(text[i: i + seq_length])
    next_chars.append(text[i + seq_length])

print(f'Number of sequences: {len(sequences)}')
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding

# Define model architecture
model = Sequential([
    tf.keras.layers.LSTM(128, input_shape=(seq_length, len(chars))),
    tf.keras.layers.Dense(len(chars), activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy')
```

```python
# Prepare input and output data for training
X = np.zeros((len(sequences), seq_length, len(chars)), dtype=np.bool)
y = np.zeros((len(sequences), len(chars)), dtype=np.bool)

for i, sequence in enumerate(sequences):
    for t, char in enumerate(sequence):
        X[i, t, char_to_index[char]] = 1
    y[i, char_to_index[next_chars[i]]] = 1

# Train the model
model.fit(X, y, batch_size=256, epochs=20)
```

```python
def generate_text(seed_text, num_chars):
    input_text = seed_text.lower()
    generated_text = seed_text

    for _ in range(num_chars):
        # Prepare the input array
        x_pred = np.zeros((1, seq_length, len(chars)))
        for t, char in enumerate(input_text):
            x_pred[0, t, char_to_index[char]] = 1

        # Predict next character
        preds = model.predict(x_pred, verbose=0)[0]
        next_index = np.argmax(preds)
        next_char = index_to_char[next_index]

        # Add the character to the generated text
        generated_text += next_char
        input_text = input_text[1:] + next_char

    return generated_text

# Generate and print text
seed_text = text[:seq_length]
generated_text = generate_text(seed_text, 400)
print("Generated Text:\n", generated_text)
```
