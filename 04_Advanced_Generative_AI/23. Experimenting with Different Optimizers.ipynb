```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Load data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize image data to [0, 1]
x_train, x_test = x_train / 255.0, x_test / 255.0

# One-hot encode labels
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

def create_model():
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(128, activation='relu'),
        Dense(10, activation='softmax')
    ])
    return model
```

```python
# Create a model instance
model_sgd = create_model()

# Compile the model
model_sgd.compile(optimizer='sgd',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

# Train the model
history_sgd = model_sgd.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

```python
# Create a new model instance
model_adam = create_model()

# Compile the model
model_adam.compile(optimizer='adam',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

# Train the model
history_adam = model_adam.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

```python
# Create a new model instance
model_rmsprop = create_model()

# Compile the model
model_rmsprop.compile(optimizer='rmsprop',
                      loss='categorical_crossentropy',
                      metrics=['accuracy'])

# Train the model
history_rmsprop = model_rmsprop.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

```python
sgd_results = model_sgd.evaluate(x_test, y_test, verbose=0)
adam_results = model_adam.evaluate(x_test, y_test, verbose=0)
rmsprop_results = model_rmsprop.evaluate(x_test, y_test, verbose=0)

print(f"SGD: Loss = {sgd_results[0]}, Accuracy = {sgd_results[1]}")
print(f"Adam: Loss = {adam_results[0]}, Accuracy = {adam_results[1]}")
print(f"RMSprop: Loss = {rmsprop_results[0]}, Accuracy = {rmsprop_results[1]}")
```

```python
import matplotlib.pyplot as plt

# Function to plot results
def plot_training_history(history, title):
    plt.plot(history.history['accuracy'], label='accuracy')
    plt.plot(history.history['val_accuracy'], label='val_accuracy')
    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_training_history(history_sgd, "SGD Optimizer")
plot_training_history(history_adam, "Adam Optimizer")
plot_training_history(history_rmsprop, "RMSprop Optimizer")
```
