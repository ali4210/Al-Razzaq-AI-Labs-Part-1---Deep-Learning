```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define a simple model
model = Sequential([
    Dense(32, activation='relu', input_shape=(784,)),
    Dense(10, activation='softmax')
])

# Convert the model architecture to JSON
model_json = model.to_json()

# Write the JSON to a file
with open("../model_architecture.json", "w") as json_file:
    json_file.write(model_json)
```

```python
# Save weights to HDF5
model.save_weights('model_weights.h5')
```

```python
from tensorflow.keras.models import model_from_json

# Load JSON and create model
with open('../model_architecture.json', 'r') as json_file:
    loaded_model_json = json_file.read()

loaded_model = model_from_json(loaded_model_json)
```

```python
# Load weights into the model
loaded_model.load_weights('model_weights.h5')
```

```python
import numpy as np

# Example input data
example_data = np.random.random((1, 784))

# Predictions with the original model
original_prediction = model.predict(example_data)

# Predictions with the loaded model
loaded_prediction = loaded_model.predict(example_data)

print("Original Prediction:", original_prediction)
print("Loaded Prediction:", loaded_prediction)
```
