```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Sample data
X_train = np.random.rand(100, 10)
y_train = np.random.rand(100, 1)

# Building the model
model_relu = Sequential([
    Dense(64, activation='relu', input_shape=(10,)),
    Dense(1)
])

# Compile the model
model_relu.compile(optimizer='adam', loss='mse')

# Training the model
model_relu.fit(X_train, y_train, epochs=10)
```

```python
# Building the model
model_sigmoid = Sequential([
    Dense(64, activation='sigmoid', input_shape=(10,)),
    Dense(1)
])

# Compile the model
model_sigmoid.compile(optimizer='adam', loss='mse')

# Training the model
model_sigmoid.fit(X_train, y_train, epochs=10)
```

```python
# Building the model
model_tanh = Sequential([
    Dense(64, activation='tanh', input_shape=(10,)),
    Dense(1)
])

# Compile the model
model_tanh.compile(optimizer='adam', loss='mse')

# Training the model
model_tanh.fit(X_train, y_train, epochs=10)
```

```python
import matplotlib.pyplot as plt

loss_values_relu = model_relu.history.history['loss']
loss_values_sigmoid = model_sigmoid.history.history['loss']
loss_values_tanh = model_tanh.history.history['loss']

plt.plot(loss_values_relu, label='ReLU')
plt.plot(loss_values_sigmoid, label='Sigmoid')
plt.plot(loss_values_tanh, label='Tanh')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss Comparison')
plt.legend()
plt.show()
```
