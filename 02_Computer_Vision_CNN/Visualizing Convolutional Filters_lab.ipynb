```python
# Using TensorFlow
from tensorflow.keras.applications import VGG16

# Load the pre-trained VGG16 model
model = VGG16(weights='imagenet', include_top=False)
```

```python
# Choosing a specific convolutional layer
layer_name = 'block1_conv1'  # Example layer name
layer_model = model.get_layer(name=layer_name)
```

```python
# Extract weights of the chosen layer
weights = layer_model.get_weights()
filters, biases = weights[0], weights[1]
```

```python
import numpy as np

# Normalize filters
f_min, f_max = filters.min(), filters.max()
filters = (filters - f_min) / (f_max - f_min)
```

```python
import matplotlib.pyplot as plt

# Define the number of filters and plot
n_filters = 6  # Number to plot
index = 1

# Plot each filter
for i in range(n_filters):
    f = filters[:, :, :, i]
    for j in range(3):  # Assuming RGB depth
        plt.subplot(n_filters, 3, index)
        plt.imshow(f[:, :, j], cmap='gray')
        plt.axis('off')
        index += 1

plt.show()
```
